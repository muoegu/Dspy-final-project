```python
chaplin_learners_all_frog = pd.read_csv('chaplin_row_data_frog.csv', sep='\t', encoding='utf-8')
chaplin_learners_all_frog.head()
```

```python
import pandas as pd
from collections import Counter
import spacy

def analyze_pos(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    word_counts = Counter()
    pos_counts = Counter()
    lemma_counts = Counter()

    for token in doc:
        word_counts[token.text] += 1
        pos_counts[token.pos_] += 1
        lemma_counts[token.lemma_] += 1

    return dict(word_counts), dict(pos_counts), dict(lemma_counts)

def add_analysis_columns(df, text_column):
    df[['word_counts', 'pos_counts', 'lemma_counts']] = df[text_column].apply(
        lambda x: pd.Series(analyze_pos(x))
    )

    return df
```

```python
chaplin_learners_all_frog = add_analysis_columns(chaplin_learners_all_frog, 'Text')
chaplin_learners_all_frog
```

```python
nlp = spacy.load('en_core_web_sm')

def add_word_pos_pairs(df, text_column):

    def get_word_pos_pairs(text):
        doc = nlp(text)
        return [(token.text, token.pos_) for token in doc]

    df['word_pos_pairs'] = df[text_column].apply(get_word_pos_pairs)

    return df
```

```python
chaplin_learners_all_frog = add_word_pos_pairs(chaplin_learners_all_frog, 'Text')
chaplin_learners_all_frog
```

```python
def add_token_details(df, text_column):
    def get_word_pos_pairs(text):
        doc = nlp(text)
        return [{'text': token.text, 'pos': token.pos_, 'lemma': token.lemma_, 'morph': token.morph} for token in doc]

    df['token_details'] = df[text_column].apply(get_word_pos_pairs)
    return df
```

```python
chaplin_learners_all_frog = add_token_details(chaplin_learners_all_frog, 'Text')
chaplin_learners_all_frog
```

```python
chaplin_learners_all_frog.to_csv('chaplin_learners_all_frog.csv', sep='\t', encoding='utf-8', index=False)
```

```python
def pre_process(file_name, test_name):
    data = pd.read_csv(file_name, sep='\t', encoding='utf-8')
    data = add_analysis_columns(data, 'Text')
    data = add_token_details(data, 'Text')
    data = add_word_pos_pairs(data, 'Text')
    data.to_csv(f'chaplin_learners_all_{test_name}.csv', sep='\t', encoding='utf-8', index=False)
```

```python
pre_process('chaplin_learners_all_famous-person_row.csv', 'famous-person')
```
